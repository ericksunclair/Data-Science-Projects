```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

# Conhecendo a base de dados

```python
df_2020_1 = pd.read_csv("data/HIST_PAINEL_COVIDBR_2020_Parte1_04fev2023.csv", sep=';')
df_2020_2 = pd.read_csv("data/HIST_PAINEL_COVIDBR_2020_Parte2_04fev2023.csv", sep=';')
df_2020 = pd.concat([df_2020_1, df_2020_2]).reset_index(drop=True)

print(df_2020.shape)
df_2020.head()
```

```python
df_2021_1 = pd.read_csv("data/HIST_PAINEL_COVIDBR_2021_Parte1_04fev2023.csv", sep=';')
df_2021_2 = pd.read_csv("data/HIST_PAINEL_COVIDBR_2021_Parte2_04fev2023.csv", sep=';')
df_2021 = pd.concat([df_2021_1, df_2021_2]).reset_index(drop=True)

print(df_2021.shape)
df_2021.head()
```

```python
df_2022_1 = pd.read_csv("data/HIST_PAINEL_COVIDBR_2022_Parte1_04fev2023.csv", sep=';')
df_2022_2 = pd.read_csv("data/HIST_PAINEL_COVIDBR_2022_Parte2_04fev2023.csv", sep=';')
df_2022 = pd.concat([df_2022_1, df_2022_2]).reset_index(drop=True)

print(df_2022.shape)
df_2022.head()
```

```python
df_completo = pd.concat([df_2020, df_2021, df_2022]).reset_index(drop=True)
print(df_completo.shape)
df_completo.head()
```

```python
for col in df_completo:
    print(col, '\n', df_completo[col].unique(), '\n')
```

```python
df_completo.isna().sum()
```

```python
print(df_completo['regiao'].value_counts())
```

O número de dados faltantes pra coluna 'estado' é o mesmo que o número linhas com a 'regiao' Brasil e o 'coduf' 76, provavelmente são as mesmas linhas. Os faltantes para 'municipio', 'codRegiaoSaude', 'nomeRegiaoSaude', 'interior/metropolitana' também devem ser os mesmos

```python
df_completo[df_completo['municipio'] == "Uruana"].nomeRegiaoSaude.unique()
```

```python
df_completo[df_completo['municipio'] == "Uruana"].estado.unique()
```

```python
df_completo[df_completo['municipio'] == "Uruana"].regiao.unique()
```

Conclusão: temos uma linha para cada dia do ano para cada município (teoricamente, pois existem dias faltantes pra cada município). Cada municipio pertence a uma única 'nomeRegiaoSaude', que pertence a um único 'estado', que pertence a uma única 'regiao'

```python
df_completo[df_completo['regiao'] == "Brasil"].nomeRegiaoSaude.unique()
```

```python
df_completo[df_completo['regiao'] == "Brasil"].estado.unique()
```

Já quando a 'regiao' é Brasil, os campos 'municipio', 'nomeRegiaoSaude', 'codRegiaoSaude' e 'estado' ficam vazios - são as 1041 ocorrências da 'regiao' Brasil.

Mas os dados faltantes para 'municipio', 'codRegiaoSaude', 'nomeRegiaoSaude' e 'interior/metropolitana' são mais: totalizam 50358 linhas e devem ser as mesmas linhas já que tem o mesmo total de dados faltantes. 

```python
df_completo[df_completo['municipio'].isna()].estado.unique()
```

No entanto, não existe padrão nesses dados faltantes, eles são de várias regiões, estados, etc. 

```python
df_completo['data'].nunique()
```

```python
for regiao in df_completo['regiao'].unique():
    print(df_completo[df_completo['regiao'] == regiao].data.nunique())
```

Existem 1041 datas únicas na base de dados, o mesmo número de ocorrências da 'regiao' Brasil. Além disso, cada 'regiao' tem também 1041 datas únicas, ou seja, todas as regiões tem seu relatório diário em cada uma das 1041 datas registradas (seja para todos os municípios dessa região, seja só pra alguns, já que nem todos os municípios tem registros pra todas as datas), além do relatório total do país que é registrado na 'regiao' Brasil.

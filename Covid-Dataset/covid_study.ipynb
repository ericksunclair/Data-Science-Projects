```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Conhecendo a base de dados

```python
df_2020_1 = pd.read_csv("data/HIST_PAINEL_COVIDBR_2020_Parte1_04fev2023.csv", sep=';')
df_2020_2 = pd.read_csv("data/HIST_PAINEL_COVIDBR_2020_Parte2_04fev2023.csv", sep=';')
df_2020 = pd.concat([df_2020_1, df_2020_2]).reset_index(drop=True)

print(df_2020.shape)
df_2020.head()
```

```python
df_2021_1 = pd.read_csv("data/HIST_PAINEL_COVIDBR_2021_Parte1_04fev2023.csv", sep=';')
df_2021_2 = pd.read_csv("data/HIST_PAINEL_COVIDBR_2021_Parte2_04fev2023.csv", sep=';')
df_2021 = pd.concat([df_2021_1, df_2021_2]).reset_index(drop=True)

print(df_2021.shape)
df_2021.head()
```

```python
df_2022_1 = pd.read_csv("data/HIST_PAINEL_COVIDBR_2022_Parte1_04fev2023.csv", sep=';')
df_2022_2 = pd.read_csv("data/HIST_PAINEL_COVIDBR_2022_Parte2_04fev2023.csv", sep=';')
df_2022 = pd.concat([df_2022_1, df_2022_2]).reset_index(drop=True)

print(df_2022.shape)
df_2022.head()
```

```python
df_completo = pd.concat([df_2020, df_2021, df_2022]).reset_index(drop=True)
print(df_completo.shape)
df_completo.head()
```

```python
for col in df_completo:
    print(col, '\n', df_completo[col].unique(), '\n')
```

```python
df_completo.isna().sum()
```

```python
df_completo[df_completo['municipio'] == "Uruana"].nomeRegiaoSaude.unique()
```

```python
df_completo[df_completo['regiao'] == "Brasil"].nomeRegiaoSaude.unique()
```

```python
df_completo[df_completo['regiao'] == "Brasil"].estado.unique()
```

Já quando a 'regiao' é Brasil, os campos 'municipio', 'nomeRegiaoSaude', 'codRegiaoSaude' e 'estado' ficam vazios - são as 1041 ocorrências da 'regiao' Brasil.

Mas os dados faltantes para 'municipio', 'codRegiaoSaude', 'nomeRegiaoSaude' e 'interior/metropolitana' são mais: totalizam 50358 linhas e devem ser as mesmas linhas já que tem o mesmo total de dados faltantes. 

```python
df_completo[df_completo['municipio'].isna()].estado.unique()
```

No entanto, não existe padrão nesses dados faltantes, eles são de várias regiões, estados, etc. 

```python
df_completo['data'].nunique()
```

Existem 1041 datas únicas na base de dados, o mesmo número de ocorrências da 'regiao' Brasil. Além disso, cada 'regiao' tem também 1041 datas únicas, ou seja, todas as regiões tem seu relatório diário em cada uma das 1041 datas registradas (seja para todos os municípios dessa região, seja só pra alguns, já que nem todos os municípios tem registros pra todas as datas), além do relatório total do país que é registrado na 'regiao' Brasil.

# Vizualizando os dados

```python
df_completo.columns
```

```python
colunas_interesse = ['regiao', 'estado', 'municipio', 'nomeRegiaoSaude', 'data', 'semanaEpi', 
                     'casosAcumulado', 'casosNovos', 'obitosAcumulado', 'obitosNovos']
df_interesse = df_completo[colunas_interesse]
df_interesse.head()
```

```python
df_semana_brasil = df_interesse[df_interesse['regiao'] == 'Brasil'].sort_values(by=['semanaEpi'])
df_semana_brasil.groupby('semanaEpi').sum()
```

A semana epidemiológica 53 parece estar incompleta (é a atual), então vou retirá-la

```python
df_semana_brasil2 = df_semana_brasil[df_semana_brasil['semanaEpi'] != 53].groupby('semanaEpi').sum()
```

```python
n_rows = 2
n_cols = 2
i = 0
colunas = df_semana_brasil2.columns

fig, axes = plt.subplots(n_rows, n_cols)
for row in range(n_rows):
    for col in range(n_cols):
        ax = axes[row][col]
        ax.plot(df_semana_brasil2[colunas[i]], marker=".")

        ax.set_title(colunas[i], fontsize=17)
        ax.set_xlabel('Semana Epidemiológica', fontsize = 12)
        ax.set_ylabel('Casos', fontsize = 12)
        ax.set_xticks(df_semana_brasil2.index)
        i += 1

fig.suptitle('Evolução dos casos ao longo das semanas epidemiológicas para o Brasil', fontsize=20)
fig.set_size_inches(30, 20)

plt.show()
```

```python
regioes = ['Norte', 'Nordeste', 'Centro-Oeste', 'Sul', 'Sudeste']

for regiao in regioes:
    df_semana = df_interesse[df_interesse['regiao'] == regiao].sort_values(by=['semanaEpi'])
    df_semana2 = df_semana[df_semana['semanaEpi'] != 53].groupby('semanaEpi').sum()
    df_semana2

    n_rows = 2
    n_cols = 2
    i = 0
    colunas = df_semana2.columns

    fig, axes = plt.subplots(n_rows, n_cols)
    for row in range(n_rows):
        for col in range(n_cols):
            ax = axes[row][col]
            ax.plot(df_semana2[colunas[i]], marker=".")

            ax.set_title(colunas[i], fontsize=17)
            ax.set_xlabel('Semana Epidemiológica', fontsize = 12)
            ax.set_ylabel('Casos', fontsize = 12)
            ax.set_xticks(df_semana2.index)
            i += 1

    fig.suptitle('Evolução dos casos ao longo das semanas epidemiológicas para a região ' + regiao, fontsize=20)
    fig.set_size_inches(30, 20)

    plt.show()
```

```python
df_semana_regiao = df_interesse[df_interesse['semanaEpi'] != 53].sort_values(by=['semanaEpi'])
df_semana_regiao = df_semana_regiao.groupby(['semanaEpi', 'regiao']).sum().reset_index()

dict_semana_regiao_wide = {}

df_semana_regiao1 = df_semana_regiao.drop(['casosAcumulado', 'obitosAcumulado', 'obitosNovos'], axis=1)
dict_semana_regiao_wide['casosNovos'] = df_semana_regiao1.pivot('semanaEpi', 'regiao', 'casosNovos')


df_semana_regiao2 = df_semana_regiao.drop(['casosNovos', 'obitosAcumulado', 'obitosNovos'], axis=1)
dict_semana_regiao_wide['casosAcumulado'] = df_semana_regiao2.pivot('semanaEpi', 'regiao', 'casosAcumulado')


df_semana_regiao3 = df_semana_regiao.drop(['casosAcumulado', 'obitosAcumulado', 'casosNovos'], axis=1)
dict_semana_regiao_wide['obitosNovos'] = df_semana_regiao3.pivot('semanaEpi', 'regiao', 'obitosNovos')


df_semana_regiao4 = df_semana_regiao.drop(['casosAcumulado', 'casosNovos', 'obitosNovos'], axis=1)
dict_semana_regiao_wide['obitosAcumulado'] = df_semana_regiao4.pivot('semanaEpi', 'regiao', 'obitosAcumulado')

dict_semana_regiao_wide['casosNovos'].head()
```

```python
dict_semana_regiao_long = {}

dict_semana_regiao_long['casosNovos'] = df_semana_regiao1
dict_semana_regiao_long['casosAcumulado'] = df_semana_regiao2
dict_semana_regiao_long['obitosNovos'] = df_semana_regiao3
dict_semana_regiao_long['obitosAcumulado'] = df_semana_regiao4

dict_semana_regiao_long['casosNovos'].head()
```

```python
colunas_valores = ['casosNovos', 'casosAcumulado', 'obitosNovos', 'obitosAcumulado']
i = 0

fig, axes = plt.subplots(nrows=4)

for coluna in colunas_valores:
    sns.lineplot(data=dict_semana_regiao_wide[colunas_valores[i]], markers=True, 
                 ax=axes[i])
    axes[i].set_title(colunas_valores[i], fontsize=17)
    axes[i].set_xlabel('Semana Epidemiológica', fontsize = 12)
    axes[i].set_ylabel('Casos', fontsize = 12)
    i += 1

fig.suptitle('Evolução dos casos ao longo das semanas epidemiológicas por região\n', fontsize=20)
fig.set_size_inches(20, 20)
fig.tight_layout()
plt.show()
```

```python
i = 0

fig, axes = plt.subplots(nrows=4)

for coluna in colunas_valores:
    sns.barplot(data=dict_semana_regiao_long[colunas_valores[i]], x='semanaEpi',
                y=colunas_valores[i], hue='regiao', ax=axes[i])
    axes[i].set_title(colunas_valores[i], fontsize=17)
    axes[i].set_xlabel('Semana Epidemiológica', fontsize = 12)
    axes[i].set_ylabel('Casos', fontsize = 12)
    i += 1

fig.suptitle('Evolução dos casos ao longo das semanas epidemiológicas por região\n', fontsize=20)
fig.set_size_inches(20, 20)
fig.tight_layout()
plt.show()
```
